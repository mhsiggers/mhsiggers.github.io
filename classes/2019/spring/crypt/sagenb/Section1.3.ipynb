<p> a % b gives us the image of a modulo b

```python
7 * 18 % 11
```


<p>Alternately, we can work in the ring of integers modulo 11.</p>

```python
M11 = Integers(11)
```

```python
M11.addition_table()
```

```python
M11(7); M11(18); M11( 7 * 18)
```

<p>To find the multiplicative inverse, we can use either of our notations:</p>
<p>&nbsp;</p>

```python
M11( 3 + 67 * 453)
```

```python
M11(7^(-1)); M11(1/7)
```


<p>We can also do modular arithmetic modulo a non-prime: 10</p>
<p>&nbsp;</p>

```python
M10=Integers(10)
```

```python
M10( 4 * 7)
```


<p>Not all integers have inverses mod 10, but some do. </p>

```python
M10.is_integral_domain()
```

```python
M10.list_of_elements_of_multiplicative_group()
```


<p> We can find the inverse of these, but now we see the distinction Sage makes between the two notations. </p>

```python
M10(1/7)
```

```python
M10(7^{-1})
```

<p> The reason it doesn't like the exponential notation is because this notation depends on its field descriptor. </p>

```python
M10.is_field(); M11.is_field()
```


<p> Observe that powering is pretty fast modulo 11.  Compare it to the 'stupid' algorithm. </p>

```python
def slowpower11(BASE,EXP):
    m = 1;
    for i in range(EXP):
        m = M11(7 * m);
    return m
```
