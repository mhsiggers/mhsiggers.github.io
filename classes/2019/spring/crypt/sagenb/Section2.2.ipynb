<p> We find a prime p and a primitive root modulo p: primroot </p>

```python
p = next_prime(200); M = Integers(p); p
```

```python
mod(313,5)
```

```python
M(217 * 12)
```

```python
primroot = M.multiplicative_generator(); primroot
```

<p> We make a discrete log function. Just the brute force one.</p>

```python
def dlog(h): 
 INDEX = 1;
 POWER = primroot;
 while POWER <> h:
   INDEX = INDEX + 1
   POWER = M(POWER * primroot);
 return INDEX
```

```python
dlog(3)
```

Check this against the sage builtin function

```python
M(3).log(primroot)
```
